```python
from jiwer import wer
import jiwer
```

```python
uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file: "{name}" with length {length} bytes'.format(name=fn, length=len(uploaded[fn])))

print(type(fn))
print(fn)
```

```python
from textgrid import TextGrid

# Load the TextGrid file
tg = TextGrid.fromFile(fn)

# Choose the tier by name (e.g., "words", "transcript", etc.)
tier_name = "words"  # Replace with the correct tier name
tier = tg.getFirst(tier_name)

print(type(tg.tiers))
print(tg.tiers)
```

```python
print(tg.tiers[0].name)

# Choose the tier by name (e.g., "words", "transcripts", etc.)
tier_name = "Doctor"
if tier_name not in [tier.name for tier in tg.tiers]:
    print(f"Tier '{tier_name}' not found in TextGrid.")
else:
    tier = tg.getFirst(tier_name)
    texts = [interval.mark for interval in tier.intervals if interval.mark.strip()]
    print("Extracted texts from tier:")
    transcript = str()
    for text in texts:
        label1 = re.search("<[A-Za-z]+>", text)
        label2 = re.search("</[A-Za-z]+>", text)
        label3 = re.search("<[A-Za-z]+/>", text)

        text = re.sub(r"<[A-Za-z]+>", "", text)
        text = re.sub(r"</[A-Za-z]+>", "", text)
        text = re.sub(r"<[A-Za-z]+/>", "", text)
        text = re.sub(r",", "", text)
        transcript += text

print(transcript)
```

```python
reference = transcript
hypothesis_MWV3 = result['text']
print(type(reference))
print(type(hypothesis_MWV3))

# MWV3
MWV3_performance = dict()
MWV3_performance['wer'] = jiwer.wer(reference, hypothesis_MWV3)
MWV3_performance['mer'] = jiwer.mer(reference, hypothesis_MWV3)
MWV3_performance['wip'] = jiwer.wip(reference, hypothesis_MWV3)
MWV3_performance['wil'] = jiwer.wil(reference, hypothesis_MWV3)
MWV3_performance['cer'] = jiwer.cer(reference, hypothesis_MWV3)
```

```python
print(MWV3_performance)
```
